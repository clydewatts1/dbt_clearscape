This will create all the UAF example tables as per documentation for 17.20 , the directory also contains the example codes

```python
import glob
import shutil
import os
from pathlib import Path
import sqlparse
import teradatasql
import yaml
```

Change directories as required

```python
# source directory
# get path of home directory
home_dir = Path.home()
# get path of source directory
src_dir = f'{home_dir}' + r'\OneDrive - Teradata\Documents\dbt_clearscape\dbt_clearscape\dbt_clearscape\utilities\UAF-Analytic-Functions-Examples_ReadMe_17-20'
# convert backslashes to forward slashes
src_dir = src_dir.replace('\\', '/')
print(src_dir)
```

```python
# get home directory
from os.path import expanduser
import pprint
home = expanduser("~")
# assume project is called dbt_clearscape
project = 'dbt_clearscape'
# read config file
with open(home + '/.dbt/profiles.yml') as file:
    config = yaml.load(file, Loader=yaml.FullLoader)
# get project config
if project not in config:
    print('project not found in config file')
project_config = config[project]
# get target
target = project_config['target']
# get connection details
project_connection = project_config['outputs'][target]
#pprint.pprint(project_connection)
```

```python
# connect to teradata
conn = teradatasql.connect(host=project_connection['server'], user=project_connection['username'], password=project_connection['password'])    
```

loop through directory and process cleanups

```python
files = glob.glob(f'{src_dir}/datasets*sql')
for file in files:
    print(file)
    # read in file 
    file_data = ""
    with open(file) as fh:
        file_data = fh.read()
    # split file into individual sql statements using sqlparse
    splits =sqlparse.split(file_data)
    # remove empty strings
    splits = [x for x in splits if x]
    # execute each sql statement
    for sql in splits:
        print(sql)
        # execute sql
        with conn.cursor() as csr:
            csr.execute("DATABASE CS01_RAW")
            csr.execute(sql, ignoreErrors=[3807,3706])

```
